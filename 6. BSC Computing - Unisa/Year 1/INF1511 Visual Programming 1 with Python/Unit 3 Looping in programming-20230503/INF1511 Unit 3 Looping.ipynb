### Write a program to print the even numbers from a start number to an end number

```python
istart=int(input('provide start number : '))
iend=int(input('provide end number : '))
if istart % 2 == 0:
    istart = istart
else:
    istart += 1
for icount in range(istart,iend,2):
    print(icount)
```

### Write a program to print the odd numbers from a start number to an end number as input from the user.

```python
istart=int(input('provide start number : '))
iend=int(input('provide end number : '))
if istart % 2 != 0:
    istart = istart
else:
    istart += 1
for icount in range(istart,iend,2):
    print(icount)
```

### Write a program that uses a FOR loop. The program must prompt the digits of your ID number.  The program must output each digit as well as the sum of the digits.

```python
inum=input('Provide your ID number : ')
numsum = 0
for num in inum:
    print(num)
    numsum = numsum + int(num)
print(f"The sum of the digits is {numsum}.")
```
